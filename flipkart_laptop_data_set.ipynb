# Now Visualized The data

```python
#Most Stars laptops
Product_popular_by_stars=data.sort_values(by='Stars',ascending=False).head(10)
Product_popular_by_stars
plt.figure(figsize=(15,8))
sns.set(style='dark')
plt.subplot(2,1,1)
ax=sns.barplot(x='Stars',y='ProductName',data=Product_popular_by_stars)
for bars in ax.containers:
    ax.bar_label(bars,fontsize=10)

```

```python
#Most Rating laptops
Product_popular_by_Ratings=data.sort_values(by='Ratings',ascending=False).head(15)
Product_popular_by_Ratings
plt.figure(figsize=(20,8))
sns.set(style='dark')
ax=sns.barplot(x='Ratings',y='ProductName',data=Product_popular_by_Ratings)
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
```

```python
#Most Review laptops
Product_popular_by_Reviews=data.sort_values(by='Reviews',ascending=False).head(15)
Product_popular_by_Reviews
plt.figure(figsize=(20,8))
sns.set(style='dark')
ax=sns.barplot(x='Reviews',y='ProductName',data=Product_popular_by_Reviews)
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
```

# Overall top best laptops

```python
overall_top_in_Proccessor=data.loc[(data['Proccessor']=='Intel Core i3 Processor (11th Gen)')^(data['Proccessor']=='Intel Core i5 Processor (11th Gen)')]
overall_top_in_Ram=overall_top_in_Proccessor.loc[(overall_top_in_Proccessor['Ram']=='8 GB DDR4 RAM')^(overall_top_in_Proccessor['Ram']=='4 GB DDR4 RAM')]
overall_top_in_Storage=overall_top_in_Ram.loc[(overall_top_in_Ram['Storage']=='512 GB SSD')^(overall_top_in_Ram['Storage']=='256 GB SSD')]
overall_top_in_current_price=overall_top_in_Storage.loc[(overall_top_in_Storage['Current_Price']<=54990)&(overall_top_in_Storage['Current_Price']>=47060)]
overall_top_in_Rating=overall_top_in_current_price.loc[(overall_top_in_current_price['Ratings']<=451)&(overall_top_in_current_price['Ratings']>=25)]
overall_top_lap=overall_top_in_Rating
overall_top_lap
```

```python

plt.figure(figsize=(25,8))
sns.set(style='dark')
ax=sns.barplot(x='ProductName',y='Current_Price',data=overall_top_lap,palette='CMRmap')
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
```
